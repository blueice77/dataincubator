```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
d = pd.read_csv("insurance.csv")
print(d.head())
```

```python
pd.DataFrame(
    {'smoking':d[d['smoker']=='yes'].groupby(['age'])['charges'].mean(),
     'nonsmoking':d[d['smoker']=='no'].groupby(['age'])['charges'].mean()}).plot.bar()
```

```python
d.hist(column='bmi', bins=100)
```

Histogram of BMI

```python
d[d['smoker']=='no'].plot(kind='scatter',x='bmi',y='charges',s=1)
```

```python
For non smokers there doesn`t appear to be a correlation between BMI and charges.
```

```python
d[d['smoker']=='yes'].plot(kind='scatter',x='bmi',y='charges',s=1)
```

Plotting charges by bmi index for smokers shows that above an BMI of around 30 the minimum charges are above 30k.
Also we see that, for smokers, below an BMI of 30 almost all of the charges are below 30k. So BMI = 30 is a threshold for increasing charges, in smokers. Conclusion: charges correlates positively with BMI, for smokers.
